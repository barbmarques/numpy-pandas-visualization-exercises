**4. Using the ```chipotle``` dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.**

````python
from env import host, password, user 

def get_db_url(user, password, host, db):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'  

url = get_db_url(user, password, host, 'chipotle')

query = """SELECT * FROM orders"""

#pd.read_sql(query, url)

orders = pd.read_sql(query, url)
````

````python
# looking at data
orders.head()
````

````python
# removing $ from item price and changing data type to float
orders['item_price'] = orders.item_price.str.replace('$','').astype(float)
orders.item_price
````

````python
#finding four most popular items
top_four = orders.groupby("item_name").quantity.sum().nlargest(4,keep='all')
top_four
````

````python
#This step corrected during walkthrough
#Using groupby and isin to calculate revenue of top four
revenue = orders[orders.item_name.isin(top_four.index)].groupby('item_name').item_price.sum()
revenue
````

````python
#Create a dataframe that I can convert to bar graph.

revenue = revenue.sort_values(ascending = False).reset_index()
revenue
````

````python
sns.barplot(y = 'item_name', x = 'item_price', data=revenue)
plt.title('Revenue from Best Sellers')

#remove column names:
plt.ylabel('')
plt.xlabel('Revenue')
````
