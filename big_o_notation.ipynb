## Big O Notation

Make a new Jupyter notebook named big_o_notation.ipynb

* Title your chart "Big O Notation"
* Label your x axis "Elements"
* Label your y axis "Operations"
* Label your curves or make a legend for the curves
* Use LaTex notation where possible

```python
import math
```

```python
import matplotlib.pyplot as plt
import math

plt.figure(figsize = (15,8))

x = range(1, 12)
y = [0 * n + 1 for n in x]
plt.plot(x,y, label = r'$\bigcirc$(1)', color = 'green')

x = range(1, 12)
y = [math.log(n) for n in x]
plt.plot(x,y, label = r'$\bigcirc$(log n)', color = 'red')

x = range(1, 12)
y = [n for n in x]
plt.plot(x,y, label = r'$\bigcirc$(n)', color = 'purple')

x = range(1, 12)
y = [n * math.log(n) for n in x]
plt.plot(x,y, label = r'$\bigcirc$(n log n)', color = 'blue')

x = range(1, 12)
y = [n ** 2 for n in x]
plt.plot(x,y, label = r'$\bigcirc$(n^2)', color = 'orange')

x = range(1, 12)
y = [2 ** n for n in x]
plt.plot(x,y, label = r'$\bigcirc$(2^n)', color = 'black')

x = range(1, 12)
y = [math.factorial(n) for n in x]
plt.plot(x,y, label = r'$\bigcirc$(n!)', color = 'sienna')

x = range(1, 12)
y = [n ** n for n in x]
plt.plot(x,y, label = r'$\bigcirc$(n^n)', color = 'magenta')

plt.xlim(1,11)
plt.ylim(0,50)

plt.xlabel('Elements')
plt.ylabel('Operations')

plt.title('Big O Notation', fontsize = 15)

plt.legend()
plt.show()


```
