1. Use matplotlib to plot the following equation:

$$y = x^2 - x + 2$$

    You'll need to write the code that generates the x and y points.

    Add an annotation for the point (0,0), the origin.

```python
import matplotlib.pyplot as plt
import numpy
import math
```

```python
# define range for x
x = range(-10, 10)

# iterate through x values to find y values
y = [ n ** 2 - n + 2 for n in x]

# call  plot of equation
plt.plot(x,y)

#add annotation
plt.annotate ('Origin', xy = (0,0), xytext = (2, 20), arrowprops = dict(facecolor = 'red'))

#add title with formatting for math text
plt.title('$y = x^2 - x + 2$', color = 'blue', fontsize = 14)

plt.show()
```

* $y = \sqrt{x}$

* $y = x^3$

* $y = 2^x$

You can use functions from the math module to help implement some of the equations above.

```python
# define range for x
x = [1, 4, 9, 16, 25, 36, 49]

# iterate through x values to find y values
y = [math.sqrt(n) for n in x]

# call  plot of equation
plt.scatter(x,y)

#add title with formatting for math text
plt.title('$y = \sqrt{x}$', fontsize = 14)

plt.show()
```

```python
x = range(-3, 3)
y = [n**3 for n in x]
plt.scatter(x,y)
plt.title('$y = x^3$', fontsize = 14)
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (10, 8))

plt.suptitle('Four Sample Equations', fontsize = 16)

plt.subplot(2,2,1)
x = [1, 4, 9, 16, 25, 36, 49]
y = [math.sqrt(n) for n in x]
plt.scatter(x,y)
plt.title('$y = \sqrt{x}$', fontsize = 14)

plt.subplot(2,2,2)
x = range(-3, 3)
y = [n**3 for n in x]
plt.scatter(x,y)
plt.title('$y = x^3$', fontsize = 14)

plt.subplot(2,2,3)
x = range(-3,3)
y = [2**n for n in x]
plt.scatter(x,y)
plt.title('$y = 2^x$', fontsize = 14)

plt.subplot(2,2,4)
x = range(0,10)
y = [1 / (n + 1) for n in x]
plt.scatter(x,y)
plt.title('$y = 1/(x+1)$', fontsize = 14)



plt.show()
```

4.  Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.


```python
plt.figure(figsize = (10, 8))

plt.suptitle('Four Sample Equations', fontsize = 16)

plt.subplot(2,2,1)
x = [1, 4, 9, 16, 25, 36, 49]
y = [math.sqrt(n) for n in x]
plt.scatter(x,y, color = 'blue', label = '$y = \sqrt{x}$')
plt.title('$y = \sqrt{x}$', fontsize = 14)
plt.legend()

plt.subplot(2,2,2)
x = range(-3, 3)
y = [n**3 for n in x]
plt.scatter(x,y, color = 'red', label = '$y = x^3$')
plt.title('$y = x^3$', fontsize = 14)
plt.legend()

plt.subplot(2,2,3)
x = range(-3,3)
y = [2**n for n in x]
plt.scatter(x,y, color = 'green', label = '$y = 2^x$')
plt.title('$y = 2^x$', fontsize = 14)
plt.legend()

plt.subplot(2,2,4)
x = range(0,10)
y = [1 / (n + 1) for n in x]
plt.scatter(x,y, color = 'm', label = '$y = 1/(x+1)$')
plt.title('$y = 1/(x+1)$', fontsize = 14)
plt.legend()

plt.show()
```
